Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

Generate Dataset

```python
n_records = 500
customer_ids = np.random.randint(1000, 1200, size=n_records)

categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Books']
product_categories = np.random.choice(categories, size=n_records)

purchase_amounts = np.round(np.random.uniform(5, 500, size=n_records), 2)

dates = pd.to_datetime('2025-01-01') + pd.to_timedelta(
    np.random.randint(0, 180, size=n_records), unit='D'
)

returning_customer = np.random.choice([0, 1], size=n_records, p=[0.6, 0.4])

df = pd.DataFrame({
    'CustomerID': customer_ids,
    'ProductCategory': product_categories,
    'PurchaseAmount': purchase_amounts,
    'TransactionDate': dates,
    'IsReturningCustomer': returning_customer
})

df.head()
```

Descriptive Statistics

```python
df['PurchaseAmount'].describe()
```

POINT 3 (Aggregation)

```python
category_summary = df.groupby('ProductCategory')['PurchaseAmount'].agg(
    Mean_Purchase='mean',
    Total_Purchase='sum'
)

category_summary
```

Histogram

```python
plt.hist(df['PurchaseAmount'], bins=30)
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Purchase Amount')
plt.show()
```

Bar Chart

```python
category_summary['Total_Purchase'].plot(kind='bar')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product Category')
plt.show()
```

This project analyzes a simulated e-commerce transaction dataset containing 500 customer purchases across four product categories: Electronics, Clothing, Home & Kitchen, and Books. Descriptive statistics show that purchase amounts vary widely, indicating a mix of low-value and high-value transactions, with the mean transaction representing moderate customer spending behavior.
